```python
import numpy as np
import pandas as pd
import random
import math
import gym
from gym import spaces
import json
from ortools.constraint_solver import pywrapcp, routing_enums_pb2
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# --------------------
# Cell 1: Load datasets (using local iot telemetry CSV)
# --------------------
# The workspace includes a telemetry CSV at: 'iot_telemetry_data 2.csv/iot_telemetry_data 2.csv'
# We'll construct a Path to that file and load it with pandas. This handles the space in the folder name.
# If you prefer a different file, update the path below.
# Ensure `data_dir` exists or set it explicitly before running this cell.
try:
    # If data_dir is already defined earlier in the notebook, keep it. Otherwise default to the notebook folder.
    data_dir
except NameError:
    data_dir = Path('.')  # notebook working dir
# Explicit path to the CSV included in the workspace (folder name contains a space)
telemetry_folder = Path('iot_telemetry_data 2.csv')
telemetry_csv = telemetry_folder / 'iot_telemetry_data 2.csv'
if telemetry_csv.exists():
    sensor_csv_path = telemetry_csv
else:
    # Fallback: try to auto-find sensor-like files in data_dir
    sensor_files = list(data_dir.glob('*sensor*.csv'))  # try to auto-find
    if sensor_files:
        sensor_csv_path = sensor_files[0]
    else:
        print('No sensor CSV found automatically. Set sensor_csv_path to your dataset.')
        sensor_csv_path = telemetry_csv  # keep target even if missing; user can adjust

# Also try to find cold chain traces (optional)
cold_files = list(data_dir.glob('*cold*csv'))
cold_csv_path = cold_files[0] if cold_files else data_dir / 'cold_chain.csv'

print('Sensor file:', sensor_csv_path)
print('Cold file:', cold_csv_path)

# Load the telemetry CSV into a DataFrame (safe read with engine fallback)
if sensor_csv_path.exists():
    try:
        df_sensors = pd.read_csv(sensor_csv_path)
    except Exception:
        df_sensors = pd.read_csv(sensor_csv_path, engine='python')
    print('Loaded telemetry CSV with shape:', df_sensors.shape)
    # display a small preview
    display(df_sensors.head())
else:
    print('Warning: telemetry CSV not found at', sensor_csv_path)
    df_sensors = pd.DataFrame()

# --------------------
# CELL 2: ML approach used -> K-means clustering + visualizations
# --------------------

# Confirm df_sensors exists
try:
    df_sensors
except NameError:
    print("df_sensors not found — ensure this cell is run after the telemetry CSV load (Cell 4).")
    raise

# Quick preview and basic cleaning
print("Raw sensor rows:", len(df_sensors))
display(df_sensors.head())

# --- 1) Select or create numeric features from telemetry ---
# Heuristics: look for common columns; adapt to your CSV column names
numeric_candidates = df_sensors.select_dtypes(include=[float, int]).columns.tolist()
print("Numeric columns detected:", numeric_candidates)

# If there are timestamp strings or lat/lon, derive features:
# Example transformations — adapt column names to your CSV
df_feat = df_sensors.copy()

# Parse timestamp if available
if 'timestamp' in df_feat.columns or 'time' in df_feat.columns:
    ts_col = 'timestamp' if 'timestamp' in df_feat.columns else 'time'
    try:
        df_feat['_ts'] = pd.to_datetime(df_feat[ts_col])
        # extract hour of day and dayofweek as features
        df_feat['_hour'] = df_feat['_ts'].dt.hour
        df_feat['_dow'] = df_feat['_ts'].dt.dayofweek
    except Exception:
        pass

# If lat/lon present create a rough "distance" feature (from depot centroid)
if 'lat' in df_feat.columns and 'lon' in df_feat.columns:
    depot_lat = df_feat['lat'].mean()
    depot_lon = df_feat['lon'].mean()
    # Haversine-like approx (not precise but ok for clustering)
    def approx_km(lat1, lon1, lat2, lon2):
        # rough conversion degrees to km
        return np.sqrt((69.0*(lat1-lat2))**2 + (69.0*(lon1-lon2)*np.cos(np.deg2rad((lat1+lat2)/2)))**2)
    df_feat['_dist_km'] = df_feat.apply(lambda r: approx_km(r['lat'], r['lon'], depot_lat, depot_lon), axis=1)

# Create some aggregate features if time-series per shipment (e.g., temp readings columns t0..tN)
# Detect columns like 't0','t1' patterns
ts_cols = [c for c in df_feat.columns if str(c).lower().startswith('t') and str(c).lstrip('t').isdigit()]
if ts_cols:
    # compute mean, max, std, time_above_4
    df_feat['_ts_mean'] = df_feat[ts_cols].mean(axis=1)
    df_feat['_ts_max'] = df_feat[ts_cols].max(axis=1)
    df_feat['_ts_std'] = df_feat[ts_cols].std(axis=1).fillna(0)
    df_feat['_degH_above_4'] = df_feat[ts_cols].apply(lambda row: np.clip(row-4.0, 0, None).sum(), axis=1)

# If there is sensor 'temp' column choose that
if 'temp' in df_feat.columns:
    df_feat['_temp'] = df_feat['temp']
elif 'temp_c' in df_feat.columns:
    df_feat['_temp'] = df_feat['temp_c']
elif 'temperature' in df_feat.columns:
    df_feat['_temp'] = df_feat['temperature']

# Now select a stable set of numeric features for clustering
preferred = ['_ts_mean','_ts_max','_ts_std','_degH_above_4','_temp','humidity','_hour','_dow','_dist_km','battery']
# keep only those present
features = [c for c in preferred if c in df_feat.columns]
if not features:
    # Fallback: use first 6 numeric columns detected earlier
    features = numeric_candidates[:6]
    print("Fallback feature set used:", features)
else:
    print("Clustering features:", features)

df_X = df_feat[features].fillna(0).astype(float)

# --- 2) Standardize features ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_X.values)

# --- 3) Choose #clusters with simple elbow method (optional) ---
inertia = []
K_range = list(range(2,8))
for k in K_range:
    kmeans_tmp = KMeans(n_clusters=k, random_state=0, n_init=10)
    kmeans_tmp.fit(X_scaled)
    inertia.append(kmeans_tmp.inertia_)

plt.figure(figsize=(6,3))
plt.plot(K_range, inertia, '-o')
plt.title("K-means elbow (inertia) — choose cluster count")
plt.xlabel("k")
plt.ylabel("inertia")
plt.grid(True)
plt.show()

# Choose k (let's pick 3 by default; you can change)
k = 3
kmeans = KMeans(n_clusters=k, random_state=0, n_init=20)
clusters = kmeans.fit_predict(X_scaled)
df_feat['_cluster'] = clusters

# --- 4) Visualize clusters with PCA 2D scatter ---
pca = PCA(n_components=2, random_state=0)
X_pca = pca.fit_transform(X_scaled)
df_feat['_pc1'] = X_pca[:,0]
df_feat['_pc2'] = X_pca[:,1]

plt.figure(figsize=(7,5))
sns.scatterplot(data=df_feat, x='_pc1', y='_pc2', hue='_cluster', palette='tab10', s=50, edgecolor='k', alpha=0.8)
plt.title(f'KMeans clusters (k={k}) projected to 2D PCA space')
plt.xlabel('PC1'); plt.ylabel('PC2')
plt.legend(title='cluster')
plt.grid(True)
plt.show()

# --- 5) Cluster centers heatmap (on original feature scale for interpretability) ---
centers = scaler.inverse_transform(kmeans.cluster_centers_)
df_centers = pd.DataFrame(centers, columns=df_X.columns)
df_centers.index.name = 'cluster'
plt.figure(figsize=(8,3 + 0.3*len(df_centers.columns)))
sns.heatmap(df_centers.T, annot=True, fmt=".2f", cmap='vlag', center=df_centers.values.mean())
plt.title('Cluster centroids (feature values)')
plt.xlabel('cluster')
plt.ylabel('feature')
plt.show()

# --- 6) Quick cluster summary stats ---
count_series = df_feat['_cluster'].value_counts().sort_index()
print("Cluster counts:")
print(count_series.to_string())

# Print a small sample per cluster
for ci in sorted(df_feat['_cluster'].unique()):
    print(f"\nSample rows for cluster {ci}:")
    display(df_feat[df_feat['_cluster']==ci].head(3)[features + ['_cluster']])

# Save clusters back to CSV for later use (optional)
out_path = Path("models") / "telemetry_with_clusters.csv"
df_feat.to_csv(out_path, index=False)
print("Saved telemetry + cluster labels to:", out_path)

# Add ML approach note for documentation
print("\nML approach used: K-means clustering (scikit-learn). Visualizations via matplotlib & seaborn. Scaling via StandardScaler.")


# --------------------
# Cell 3: Simple spoilage model & helper (domain knowledge)
# --------------------
# We'll use a simplified spoilage model: each product has a nominal 'safe_temp_range' (min,max).
# Each hour the product spends above safe_temp increases 'spoilage_score'. If score > threshold -> spoiled.
# We will create synthetic product temperature thresholds (in real usage, load from dataset / FSIS FoodKeeper).

# Example: dictionary of product types and safe temps (Celsius). Real lists like FoodKeeper exist.
product_safe_temp = {
    "milk": (0.0, 4.0),
    "fresh_fish": (0.0, 2.0),
    "lettuce": (0.5, 5.0),
    "chicken": (0.0, 4.0)
}

def spoilage_increment(temp, safe_range):
    """Return spoilage increment per hour given measured temp and safe_range tuple (min,max)."""
    tmin, tmax = safe_range
    if tmin <= temp <= tmax:
        return 0.0
    # linear penalty for deviation magnitude (hours-weighted)
    dev = max(temp - tmax, tmin - temp)
    # exponentiate a bit to reflect faster spoilage at higher deviations
    return 0.01 * (1 + dev**1.2)

# --------------------
# Cell 4: A small VRP solver helper using OR-Tools
# --------------------
def solve_vrp(distance_matrix, vehicle_capacities, demands, num_vehicles):
    """Return routes (list of lists of node indices) and total distance using OR-Tools CVRP solver."""
    n = len(distance_matrix)
    manager = pywrapcp.RoutingIndexManager(n, num_vehicles, 0)
    routing = pywrapcp.RoutingModel(manager)

    # Distance callback
    def distance_callback(from_index, to_index):
        # Cast OR-Tools provided indices to plain Python ints to avoid overflow/conversion issues
        from_index = int(from_index)
        to_index = int(to_index)
        from_node = manager.IndexToNode(from_index)
        to_node = manager.IndexToNode(to_index)
        # Ensure the matrix element is a Python float before scaling
        return int(float(distance_matrix[from_node][to_node]) * 1000)  # scale to integer

    transit_callback_index = routing.RegisterTransitCallback(distance_callback)
    routing.SetArcCostEvaluatorOfAllVehicles(transit_callback_index)

    # Add capacity constraint
    def demand_callback(from_index):
        # Cast index to plain Python int as above
        from_index = int(from_index)
        from_node = manager.IndexToNode(from_index)
        return int(float(demands[from_node]))

    demand_callback_index = routing.RegisterUnaryTransitCallback(demand_callback)
    # Ensure vehicle capacities are plain ints (one per vehicle)
    vehicle_capacities = [int(v) for v in vehicle_capacities]
    routing.AddDimensionWithVehicleCapacity(
        demand_callback_index,
        0,  # null capacity slack
        vehicle_capacities,
        True,
        "Capacity"
    )

    # Search parameters
    search_parameters = pywrapcp.DefaultRoutingSearchParameters()
    search_parameters.first_solution_strategy = (
        routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC)
    search_parameters.time_limit.seconds = 2

    solution = routing.SolveWithParameters(search_parameters)
    routes = []
    total_dist = 0
    if solution:
        for v in range(num_vehicles):
            idx = routing.Start(v)
            route = []
            while not routing.IsEnd(idx):
                node = manager.IndexToNode(idx)
                route.append(node)
                previous_index = idx
                idx = solution.Value(routing.NextVar(idx))
                total_dist += routing.GetArcCostForVehicle(previous_index, idx, v) / 1000.0
            routes.append(route)
    else:
        # fallback: no solution -> return greedy single routes
        for v in range(num_vehicles):
            routes.append([])
    return routes, total_dist

# --------------------
# Cell 5: Define Gym environment: ColdChainEnv
# --------------------
class ColdChainEnv(gym.Env):
    """
    Observation:
      - inventory_levels for each product at depot (vector)
      - current temperatures in each vehicle (vector)
      - time_of_day (scaled)
      - pending customer demands (aggregate or small vector)
    Actions:
      - Replenish decisions (order quantities per product, discretized)
      - Temperature setpoint per vehicle (discrete choices)
      - Routing assignment choice (we'll let an internal VRP solver compute routes, but agent chooses a "fleet size" or "routing aggressiveness" parameter)
    Reward:
      - negative spoilage penalty, negative cost (transport cost + refrigeration energy), on-time delivery penalty, positive reward for meeting demand.
    """
    metadata = {"render.modes": ["human"]}

    def __init__(self,
                 product_list=["milk","lettuce"],
                 initial_inventory=None,
                 vehicle_count=3,
                 vehicle_capacity=100,
                 time_horizon=24,
                 orders_per_step=5):
        super().__init__()
        self.products = product_list
        self.P = len(self.products)
        self.vehicle_count = vehicle_count
        self.vehicle_capacity = [vehicle_capacity] * vehicle_count
        self.time_horizon = time_horizon
        self.t = 0  # current hour
        self.orders_per_step = orders_per_step

        # state arrays
        self.inventory = np.array([initial_inventory.get(p, 500) if initial_inventory else 500 for p in self.products], dtype=np.float32)
        self.temp_in_vehicles = np.array([4.0] * vehicle_count, dtype=np.float32)  # Celsius
        # pending customer demands: we'll create some synthetic demands
        self.pending_demands = np.zeros(self.P, dtype=np.float32)

        # action space: for simplicity, discretize:
        # - replenish multiplier per product: {0, 1 (small), 2 (medium), 3 (large)} -> P values
        # - temp setpoint choice per vehicle: {1:2C, 2:4C, 3:6C}
        # - routing aggression: {0,1,2} (0 = consolidate to fewer vehicles, 2 = use more vehicles -> affects emissions/cost)
        replen_actions = 4
        temp_actions = 3
        routing_actions = 3

        self.action_space = spaces.MultiDiscrete([replen_actions]*self.P + [temp_actions]*vehicle_count + [routing_actions])
        # observation: inventory(P), temps(vehicle_count), pending(P), time_of_day (1 value)
        obs_len = self.P + vehicle_count + self.P + 1
        self.observation_space = spaces.Box(low=-10000, high=10000, shape=(obs_len,), dtype=np.float32)

        # Spoilage trackers per product (0 to 1)
        self.spoilage_score = np.zeros(self.P, dtype=np.float32)

        # static distances for VRP: create synthetic depot + customers positions
        self.num_customers = 8
        self.locations = np.random.rand(self.num_customers+1, 2) * 100  # including depot at index 0
        # demands per customer (by product aggregated) -> generate numeric demands
        self.customer_demands = np.random.randint(1, 20, size=(self.num_customers, ))

        # cost params
        self.transport_cost_per_km = 1.0
        self.refrigeration_energy_per_degree_per_km = 0.01  # simplified
        self.spoilage_penalty = 100.0

        # product safe temps
        self.safe_ranges = [product_safe_temp.get(p,(0.0,4.0)) for p in self.products]

    def reset(self):
        self.t = 0
        self.inventory = np.array([500.0]*self.P, dtype=np.float32)
        self.temp_in_vehicles = np.array([4.0]*self.vehicle_count, dtype=np.float32)
        self.pending_demands = np.array([10.0]*self.P, dtype=np.float32)
        self.spoilage_score = np.zeros(self.P, dtype=np.float32)
        self.locations = np.random.rand(self.num_customers+1, 2) * 100
        self.customer_demands = np.random.randint(1, 20, size=(self.num_customers, ))
        return self._get_obs()

    def _get_obs(self):
        return np.concatenate([self.inventory, self.temp_in_vehicles, self.pending_demands, np.array([self.t / self.time_horizon])]).astype(np.float32)

    def step(self, action):
        # parse action vector
        idx = 0
        # replenish for each product
        replen_multipliers = []
        for p in range(self.P):
            m = int(action[idx]); idx += 1
            replen_multipliers.append(m)
            # map multiplier to absolute order
            order_qty = m * 50.0  # 0,50,100,150
            self.inventory[p] += order_qty

        # temp setpoints for each vehicle
        temp_choices = []
        for v in range(self.vehicle_count):
            choice = int(action[idx]); idx += 1
            # map 0->2C,1->4C,2->6C
            setpoint = 2.0 if choice == 0 else (4.0 if choice == 1 else 6.0)
            temp_choices.append(setpoint)
            self.temp_in_vehicles[v] = setpoint  # assume immediate set

        # routing parameter
        routing_choice = int(action[idx]); idx += 1
        # map routing_choice to number of vehicles to use (0->use fewer;2->use all)
        vehicles_to_use = max(1, int(self.vehicle_count * (0.5 + 0.25*routing_choice)))

        # Build distance matrix for current synthetic locations
        coords = self.locations
        n = len(coords)
        dist = np.zeros((n,n))
        for i in range(n):
            for j in range(n):
                dist[i,j] = np.linalg.norm(coords[i]-coords[j])

        # demands vector for OR-Tools: include depot at index 0
        demands = np.concatenate(([0], self.customer_demands))
        # Solve VRP quickly
        routes, total_dist = solve_vrp(dist.tolist(), [self.vehicle_capacity[0]]*vehicles_to_use, demands.tolist(), vehicles_to_use)

        # Compute transport cost and refrigeration energy
        transport_cost = total_dist * self.transport_cost_per_km
        # refrigeration energy: sum over vehicles of (|setpoint - ambient| * distance)
        ambient = 20.0
        avg_setpoint = np.mean(temp_choices)
        refrigeration_energy = total_dist * abs(avg_setpoint - ambient) * self.refrigeration_energy_per_degree_per_km

        # Meet demands: reduce pending demands and inventory
        delivered = min(self.inventory.sum(), self.pending_demands.sum())
        # For simplicity assume we meet proportional demand by product
        if self.pending_demands.sum() > 0:
            proportions = self.pending_demands / self.pending_demands.sum()
        else:
            proportions = np.zeros_like(self.pending_demands)
        delivered_per_product = delivered * proportions
        self.inventory -= delivered_per_product
        self.pending_demands = np.maximum(0.0, self.pending_demands - delivered_per_product)

        # Update spoilage score based on vehicle temps and time spent (1 hour per step)
        for p in range(self.P):
            # assume portion of that product is in vehicles at avg_setpoint
            temp = avg_setpoint
            incr = spoilage_increment(temp, self.safe_ranges[p])
            self.spoilage_score[p] += incr
            # If spoilage_score passes threshold, those units considered spoiled:
            if self.spoilage_score[p] > 1.0:
                # proportion spoiled of current inventory
                spoiled_amount = 0.1 * self.inventory[p]
                self.inventory[p] = max(0.0, self.inventory[p] - spoiled_amount)

        # Compute penalties & reward
        spoilage_penalty = self.spoilage_score.sum() * self.spoilage_penalty
        cost = transport_cost + refrigeration_energy
        on_time_penalty = 0.0
        # simple on-time: penalize if pending_demands remain large
        on_time_penalty = 50.0 * (self.pending_demands.sum() / (len(self.pending_demands) * 10.0))

        reward = - (spoilage_penalty + cost + on_time_penalty)

        # increment time
        self.t += 1
        done = (self.t >= self.time_horizon)

        info = {
            "transport_cost": transport_cost,
            "refrigeration_energy": refrigeration_energy,
            "spoilage_penalty": spoilage_penalty,
            "pending_demands": float(self.pending_demands.sum())
        }

        return self._get_obs(), float(reward), done, info

    def render(self, mode="human"):
        print(f"t={self.t}, inventory={self.inventory}, temps={self.temp_in_vehicles}, spoilage={self.spoilage_score}, pending={self.pending_demands}")

# --------------------
# Cell 6: Create the environment and wrap for SB3
# --------------------
env = ColdChainEnv(product_list=["milk","fresh_fish","lettuce"], vehicle_count=3, time_horizon=48)
vec_env = DummyVecEnv([lambda: env])

# --------------------
# Cell 7: Train RL agent (PPO)
# --------------------
model = PPO("MlpPolicy", vec_env, verbose=1, batch_size=64, learning_rate=3e-4)
# Training steps: adjust as necessary (this is an example)
train_timesteps = 5000
model.learn(total_timesteps=train_timesteps)

# Save model
model.save("coldchain_ppo_model")

# --------------------
# Evaluation: run a few episodes with the trained policy to create df_eval
# --------------------
# Create a fresh (non-vectorized) environment for evaluation so we can step and collect scalar rewards
eval_env = ColdChainEnv(product_list=["milk","fresh_fish","lettuce"], vehicle_count=3, time_horizon=48)

n_eval_episodes = 10
rewards = []
for ep in range(n_eval_episodes):
    obs = eval_env.reset()
    done = False
    ep_reward = 0.0
    while not done:
        # model.predict accepts single observation arrays for non-vec envs
        action, _ = model.predict(obs, deterministic=True)
        obs, reward, done, info = eval_env.step(action)
        ep_reward += reward
    rewards.append(ep_reward)

# Create df_eval the plotting cell expects
if 'df_eval' in globals():
    # preserve existing df_eval if present (do not overwrite)
    pass
else:
    df_eval = pd.DataFrame({"reward": rewards})
print('Evaluation rewards (per episode):')
print(df_eval)

# --------------------
# Cell 8: Plot basic metrics
# --------------------
plt.figure(figsize=(8,4))
if 'df_eval' not in globals():
    print("df_eval not found. Creating placeholder evaluation results.")
    rewards = [float(np.random.normal(loc=0.0, scale=1.0)) for _ in range(10)]
    df_eval = pd.DataFrame({"reward": rewards})
    print(df_eval)
plt.plot(df_eval["reward"].values)
plt.title("Episode rewards")
plt.xlabel("episode")
plt.ylabel("reward")
plt.grid(True)
plt.show()
```

## ML approach used
This notebook applies K-means clustering (scikit-learn) to telemetry data to discover operational clusters.\n
- Clustering: K-means (scikit-learn)\n
- Feature scaling: StandardScaler (scikit-learn)\n
- Dimensionality reduction for visualization: PCA (scikit-learn)\n
- Visualizations: matplotlib and seaborn\n

The clustering cell also computes an elbow plot to help choose k, projects clusters to 2D with PCA, and shows cluster centroids via a heatmap for interpretability.
